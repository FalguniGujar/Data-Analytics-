##**Chart-6**

##**Question no.6 What percentage of total bookings were cancelled? **

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('/content/Hotel Bookings (1).csv')

# Calculate cancellation and confirmation rates for both hotel types
cancellation_data = df.groupby('hotel')['is_canceled'].value_counts(normalize=True).unstack() * 100
cancellation_data.columns = ['Confirmed', 'Canceled']  # Rename columns for clarity

# Plot stacked bar chart
cancellation_data.plot(kind='bar', stacked=True, figsize=(8, 5), color=['darkblue', 'yellow'])

# Labels and title
plt.xlabel("Hotel Type")
plt.ylabel("Percentage (%)")
plt.title("Booking Cancellation & Confirmation Rate for City and Resort Hotels")
plt.legend(["Confirmed", "Canceled"], loc="upper right")
plt.xticks(rotation=0)

# Show plot
plt.show()

# Display exact percentage values
print("Booking Cancellation & Confirmation Rate (%) for each hotel type:")
print(cancellation_data)
```

## ***1. Why did you pick the specific chart?***

A bar chart effectively compares the proportion of canceled and non-canceled bookings, providing a clear visual representation of booking trends.


## ***2. What is/are the insight(s) found from the chart?***

Most of the bookings were not canceled. However, there is a significant number of cancellations, which could indicate customer dissatisfaction, pricing issues, or external factors.


## ***3. Will the gained insights help creating a positive business impact?***

Yes. By understanding cancellation rates, hotels can analyze the reasons behind them and implement policies like flexible cancellation, better pricing strategies, or loyalty programs to reduce cancellations.

## **Question no. 7 Which month had the highest number of hotel bookings **

```python
# Convert arrival_date_month to categorical for correct ordering
df['arrival_date_month'] = pd.Categorical(df['arrival_date_month'],
categories=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'],
ordered=True)

# Count plot for bookings per month
plt.figure(figsize=(10, 6))
sns.countplot(x='arrival_date_month', data=df, palette='Blues_d')

# Labels and title
plt.xlabel("Month")
plt.ylabel("Number of Bookings")
plt.title("Busiest Month for Hotels")

plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

## ***1. Why did you pick the specific chart?***

A bar chart is ideal for showing variations in booking trends over different months, making it easier to identify peak and off-peak seasons.

## ***2. What is/are the insight(s) found from the chart?***

The highest number of bookings occurred in October, with other peaks from April to June. The lowest bookings happened at the start and end of the year.


## ***3. Will the gained insights help creating a positive business impact?***
Yes. The hotel can use this insight to plan marketing campaigns, adjust room pricing, and offer discounts during off-peak seasons to increase occupancy rates.
